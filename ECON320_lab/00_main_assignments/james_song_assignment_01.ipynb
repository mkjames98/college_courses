1). Create an object "birthday" which is an integer formatted as "DDMMYYYY". For example, if your birthday is December 19 of 1990, then ```birthday = 19121990```, use this object as your seed in the following two exercises. 

````python
# feb 4th 1998
birthday = "04021998"
# set seed
rng = np.random.RandomState(int(birthday))
````

2). Write a script that flips a fair coin 10 times and generates a mean when heads = 1 and tails = 0. Then perform that process 1000 times, plot the distribution of those means, and report the mean of the means and the standard deviation of the means. 

````python
def gen_mean_n_flip(birthday_string: str, rng: np.random.mtrand.RandomState, n: int) -> np.float64:
    cointoss = rng.choice(2, n) # takes n flips as input
    # 2 is the result (1 or 0) for cointoss
    # 10 is the amount of times this process is repeated.
    return np.mean(cointoss)

def gen_mean_1k_trials(birthday_string, rng, n) -> list:
    # list comprehension
    # mean of tossing coin toss trial (1 trial = cointoss 10 times) x 1000 saved as list
    thousand_means = [gen_mean_n_flip(birthday_string, rng, n) for _ in range(0,1000)]
    return thousand_means
````

````python
means_10x10000 = gen_mean_1k_trials(birthday, rng, 10) 
````

````python
mean_distribution_plot_10x1000 = sns.displot(
    data=means_10x10000,
    bins=10,
    stat="density",
    kde=True,
    color="red",
    line_kws={"lw": 4},
    facecolor="blue",
    edgecolor="white",
    kind="hist",
)

plt.xlabel("Mean of Each Trial")
plt.title("distribution of 10 x 1000 coin toss trials")
print("With the seed set to my birthday,")
print("And 10 flips each trial, repeated for 1000 times,")
print(f"The Mean of the means is: {np.mean(means_10x10000)}")
print(f"The Standard Deviation of the means is: {np.std(means_10x10000)}")
print("\nThe distribution is plotted below:")
plt.show()
````

3). Write a script that flips a fair coin 100 times and generates a mean, as above. Then perform that process 1000 times, plot the distribution of those means and report the mean of the means and the standard deviation of the means.

````python
means_100x10000 = gen_mean_1k_trials(birthday, rng, 100)
mean_distribution_plot_100x1000 = sns.displot(
    data=means_100x10000,
    bins=10,
    stat="density",
    kde=True,
    color="red",
    line_kws={"lw": 4},
    facecolor="darkblue",
    edgecolor="black",
    kind="hist",
)

plt.xlabel("Mean of Each Trial")
plt.title("distribution of 100 x 1000 coin toss trials")
print("With the seed set to my birthday,")
print("And 100 flips each trial, repeated for 1000 times,")
print(f"The Mean of the means is: {np.mean(means_100x10000)}")
print(f"The Standard Deviation of the means is: {np.std(means_100x10000)}")
print("\nThe distribution is plotted below:")
plt.show()
````
